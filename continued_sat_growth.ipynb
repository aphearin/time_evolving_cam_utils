## now match at accretion

```python
from astropy.cosmology import Planck13
```

```python
histories['halo_scale_factor_lastacc'][histories['halo_scale_factor_lastacc']>1] = 1.
nonzero_mask = histories['halo_scale_factor_lastacc']!=0.
histories = histories[nonzero_mask]
ztemp = 1./histories['halo_scale_factor_lastacc']-1
histories['lookback_time_of_accretion'] = Planck13.lookback_time(ztemp)
```

```python
satmask_acc_match = (histories['upid'] != -1) & (histories['lookback_time_of_accretion'] < 5.5)
satmask_acc_match *= histories['lookback_time_of_accretion'] > 4.5
satmask_acc_match *= histories['sm_5gyr_ago'] > 10**9.8
satmask_acc_match *= histories['sm_5gyr_ago'] < 10**10.
satellites_acc_match = histories[satmask_acc_match]
print(len(satellites_acc_match))

cenmask_acc_match = (histories['upid'] == -1) 
cenmask_acc_match *= histories['sm_5gyr_ago'] > 10**9.8
cenmask_acc_match *= histories['sm_5gyr_ago'] < 10**10.
centrals_acc_match = histories[cenmask_acc_match]
print(len(cenmask_acc_match))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

__ = ax.hist(np.log10(satellites_acc_match['sm']), bins=np.linspace(9.5, 10.5, 50), normed=True, 
              color='red', label=r'satellites $\log_{10}M_{*}(z=z_{\rm acc}=0.5) \approx 10$', alpha=0.5)
__ = ax.hist(np.log10(centrals_acc_match['sm']), bins=np.linspace(9.5, 10.5, 50), normed=True, 
              color='blue', label=r'centrals $\log_{10}M_{*}(z=z^{\rm sats}_{\rm acc}=0.5) \approx 10$', alpha=0.5)

plt.xlim(xmin=9.75, xmax=10.3)
plt.ylim(ymin=0, ymax=11)
xlabel = ax.set_xlabel(r'$\log_{10}M_{*}(z=0)$', size=17)
ylabel = ax.set_ylabel('PDF', size=18)
plt.legend(frameon=False, loc='best', fontsize=16)

fig.savefig('FIGS/zacc_match_sm_comparison.pdf', bbox_extra_artists=[xlabel, ylabel], bbox_inches='tight')
```
