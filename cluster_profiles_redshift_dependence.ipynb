# Load the pre-computed data

```python
from halotools.mock_observables import return_xyz_formatted_array, radial_profile_3d

sample_z0 = Table.read('PLOT_DATA/logsm9_z0.hdf5', path='data')
sample_z1 = Table.read('PLOT_DATA/logsm9_z1.hdf5', path='data')
sample_z2 = Table.read('PLOT_DATA/logsm9_z2.hdf5', path='data')
sample_z0_pos = return_xyz_formatted_array(sample_z0['x'], sample_z0['y'], sample_z0['z'], period=250)
sample_z1_pos = return_xyz_formatted_array(sample_z1['x'], sample_z1['y'], sample_z1['z'], period=250)
sample_z2_pos = return_xyz_formatted_array(sample_z2['x'], sample_z2['y'], sample_z2['z'], period=250)

clusters_z0 = Table.read('PLOT_DATA/clusters_z0.hdf5', path='data')
clusters_z1 = Table.read('PLOT_DATA/clusters_z1.hdf5', path='data')
clusters_z2 = Table.read('PLOT_DATA/clusters_z2.hdf5', path='data')
clusters_z0_pos = return_xyz_formatted_array(clusters_z0['x'], clusters_z0['y'], clusters_z0['z'], period=250)
clusters_z1_pos = return_xyz_formatted_array(clusters_z1['x'], clusters_z1['y'], clusters_z1['z'], period=250)
clusters_z2_pos = return_xyz_formatted_array(clusters_z2['x'], clusters_z2['y'], clusters_z2['z'], period=250)
```

```python
rbins_normalized = np.linspace(0.25, 25, 20)
rbins_normalized_midpoints = (rbins_normalized[:-1] + rbins_normalized[1:])/2.

sample_z0_profile = radial_profile_3d(clusters_z0_pos, sample_z0_pos, sample_z0['ssfr'],
                                      normalize_rbins_by = clusters_z0['rvir']/1000.,
                                      rbins_normalized = rbins_normalized, period=250)

sample_z1_profile = radial_profile_3d(clusters_z1_pos, sample_z1_pos, sample_z1['ssfr'],
                                      normalize_rbins_by = clusters_z1['rvir']/1000.,
                                      rbins_normalized = rbins_normalized, period=250)
```

```python
from scipy.signal import savgol_filter, medfilt

fig = plt.figure()
ax = fig.add_subplot(111)

y1 = medfilt(sample_z1_profile, 3)
y1 = savgol_filter(y1, 9, 3)
ax.plot(rbins_normalized_midpoints, y1 - y1.max(), 
        color='green', linewidth=2, label='z=1')

y0 = medfilt(sample_z0_profile, 3)
y0 = savgol_filter(y0, 9, 3)
ax.plot(rbins_normalized_midpoints, y0 - y0.max(), 
        color='blue', linewidth=2, label='z=0')

ax.plot(np.arange(100), np.zeros(100), '--', color='gray')
xlabel = ax.set_xlabel(r'$r / R_{\rm vir}$', fontsize=16)
ylabel = ax.set_ylabel(r'$\delta\langle{\rm sSFR}\rangle$', fontsize=16)
title = ax.set_title(r'Normalized SFR profiles around clusters', 
                     fontsize=13, y=1.02) 

plt.xticks(fontsize=11); plt.yticks(fontsize=11)
plt.xlim(xmin = 0., xmax=14)
plt.ylim(ymin = -0.2, ymax = 0.0)

plt.legend(loc='center right', fontsize=13)

fig.savefig('FIGS/cluster_profiles_normalized_z_dependence.pdf', 
            bbox_extra_artists=[xlabel, ylabel], bbox_inches='tight')
```

## Galaxy type fraction profiles

```python
frac_true_cen_profile_z0 = radial_profile_3d(clusters_z0_pos, sample_z0_pos, sample_z0['gal_type']=='central',
                                             normalize_rbins_by = clusters_z0['rvir']/1000.,
                                             rbins_normalized = rbins_normalized, period=250)

frac_backsplash_profile_z0 = radial_profile_3d(clusters_z0_pos, sample_z0_pos, sample_z0['gal_type']=='backsplash',
                                               normalize_rbins_by = clusters_z0['rvir']/1000., 
                                               rbins_normalized = rbins_normalized, period=250)
```

```python
frac_true_cen_profile_z1 = radial_profile_3d(clusters_z1_pos, sample_z1_pos, sample_z1['gal_type']=='central',
                                             normalize_rbins_by = clusters_z1['rvir']/1000.,
                                             rbins_normalized = rbins_normalized, period=250)

frac_backsplash_profile_z1 = radial_profile_3d(clusters_z1_pos, sample_z1_pos, sample_z1['gal_type']=='backsplash',
                                               normalize_rbins_by = clusters_z1['rvir']/1000., 
                                               rbins_normalized = rbins_normalized, period=250)
```

```python
from scipy.signal import savgol_filter, medfilt

fig = plt.figure()
ax = fig.add_subplot(111)

y0a = medfilt(frac_true_cen_profile_z0, 3)
y0a = savgol_filter(y0a, 9, 3)
ax.plot(rbins_normalized_midpoints, 1-y0a, 
        color='blue', linewidth=2, label='z=0')

y1a = medfilt(frac_true_cen_profile_z1, 3)
y1a = savgol_filter(y1a, 9, 3)
ax.plot(rbins_normalized_midpoints, 1-y1a, 
        color='green', linewidth=2, label='z=1')

xlabel = ax.set_xlabel(r'$r / R_{\rm vir}$', size=16)
ylabel = ax.set_ylabel(r'satellite fraction', size=15)
title = ax.set_title(r'$M_{\ast}(z) = 10^{9}M_{\odot}$', fontsize=16, y=1.02) 

plt.xticks(fontsize=11); plt.yticks(fontsize=11)
plt.xlim(xmin = 0., xmax=14)
plt.ylim(ymin = 0, ymax = 1)

plt.legend(loc='best', fontsize=15)
fig.savefig('FIGS/cluster_profiles_satellite_fraction.pdf', 
            bbox_extra_artists=[xlabel, ylabel], bbox_inches='tight')
```
